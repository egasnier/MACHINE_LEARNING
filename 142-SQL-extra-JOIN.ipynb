# INNER JOIN = intersection
Il s'agit de la jointure la plus simple, la mieux comprise et la plus courante.     
Cette requte renverra tous les enregistrements de la table de gauche (table A) qui ont un enregistrement correspondant dans la table de droite (table B). Cette jointure s'crit comme suit :


```python
SELECT <select_list> 
FROM Table_A A
INNER JOIN Table_B B
ON A.Key = B.Key
```

# LEFT JOIN = rejoindre  gauche
Cette requte renverra tous les enregistrements de la table de gauche (table A), que l'un de ces enregistrements ait ou non une correspondance dans la table de droite (table B).    
Il renverra galement tous les enregistrements correspondants de la table de droite. Cette jointure s'crit comme suit :

```python
SELECT <select_list>
FROM Table_A A
LEFT JOIN Table_B B
ON A.Key = B.Key
```

# RIGHT JOIN = rejoindre  droite
Cette requte renverra tous les enregistrements de la table de droite (table B), que l'un de ces enregistrements ait ou non une correspondance dans la table de gauche (table A). Il renverra galement tous les enregistrements correspondants de la table de gauche.   
Cette jointure s'crit comme suit :

```python
SELECT <select_list>
FROM Table_A A
RIGHT JOIN Table_B B
ON A.Key = B.Key
```

# OUTER JOIN = jointure externe
Cette jointure peut galement tre appele un FULL OUTER JOINou un FULL JOIN. Cette requte renverra tous les enregistrements des deux tables, en joignant les enregistrements de la table de gauche (table A) qui correspondent aux enregistrements de la table de droite (table B).    
Cette jointure s'crit comme suit :

```python
SELECT <select_list>
FROM Table_A A
FULL OUTER JOIN Table_B B
ON A.Key = B.Key
```

# LEFT JOIN WHERE = gauche hors JOIN
Cette requte renverra tous les enregistrements de la table de gauche (table A) qui ne correspondent  aucun enregistrement de la table de droite (table B).    
Cette jointure s'crit comme suit :

```python
SELECT <select_list> 
FROM Table_A A
LEFT JOIN Table_B B
ON A.Key = B.Key
WHERE B.Key IS NULL
```

# RIGHT JOIN WHERE = droit hors JOIN
Cette requte renverra tous les enregistrements de la table de droite (table B) qui ne correspondent  aucun enregistrement de la table de gauche (table A).   
Cette jointure s'crit comme suit :

```python
SELECT <select_list>
FROM Table_A A
RIGHT JOIN Table_B B
ON A.Key = B.Key
WHERE A.Key IS NULL
```

# FULL OUTER JOIN WHERE = externe excluant JOIN
Cette requte renverra tous les enregistrements de la table de gauche (table A) et tous les enregistrements de la table de droite (table B) qui ne correspondent pas.    
Je n'ai pas encore eu besoin d'utiliser ce type de jointure, mais j'utilise assez frquemment tous les autres. Cette jointure s'crit comme suit :

```python
SELECT <select_list>
FROM Table_A A
FULL OUTER JOIN Table_B B
ON A.Key = B.Key
WHERE A.Key IS NULL OR B.Key IS NULL
```
